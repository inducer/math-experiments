```python
ts = [0]
xs = [10]
vs = [15]

g = 9.81
x0 = 5
m = 1
dt = 0.001
k = 1

for i in range(100000):
    t = ts[-1]
    v = vs[-1]
    x = xs[-1]
    
    xnext = x + dt*v
    vnext = v + dt*(g - k*(x-x0)/m -0.03*v)
    tnext = t + dt
    
    ts.append(tnext)
    xs.append(xnext)
    vs.append(vnext)
```

```python
import matplotlib.pyplot as plt

plt.plot(ts, xs)
```
